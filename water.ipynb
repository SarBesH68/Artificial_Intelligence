Jug X capacity = 4 Liters    
Jug Y capacity = 3 Liters    
Initial State = (0,0)     
Goal State = (2,0)    

Allowed Operators:    
1.Fill Jug X    
2.Fill Jug Y     
3.Empty Jug X    
4.Empty Jug Y    
5.Pour X->Y   
6.Pour Y->X   

```python
from collections import deque

x_capacity = int(input("Enter capacity of Jug X: "))
y_capacity = int(input("Enter capacity of Jug Y: "))

goal_x = int(input("Enter desired amount in Jug X: "))
goal_y = int(input("Enter desired amount in Jug Y: "))

start = (0, 0)
goal = (goal_x, goal_y)


def get_next_states(state, x_capacity, y_capacity):
    x, y = state
    states = []

    # Fill X
    states.append((x_capacity, y))

    # Fill Y
    states.append((x, y_capacity))

    # Empty X
    states.append((0, y))

    # Empty Y
    states.append((x, 0))

    # Pour X -> Y
    transfer = min(x, y_capacity - y)
    states.append((x - transfer, y + transfer))

    # Pour Y -> X
    transfer = min(y, x_capacity - x)
    states.append((x + transfer, y - transfer))

    return states


def bfs(start, goal, x_capacity, y_capacity):

    visited = set([start]) #for node which are already visited

    queue = deque([[start]]) # nodes in queue

    while queue: #runs until the queue is empty

        path = queue.popleft() #first element is popped and stored in path variable

        current = path[-1] #last value of path is stored at current variable

        if current == goal: #if the current path is goal state then print solution is found
            print("Solution Found\n")

            for state in path: #print path
                print(state)

            return

        for neighbour in get_next_states(current, x_capacity, y_capacity):

            if neighbour not in visited:

                visited.add(neighbour)

                queue.append(path + [neighbour])

    print("No solution found.")


bfs(start, goal, x_capacity, y_capacity)
```
